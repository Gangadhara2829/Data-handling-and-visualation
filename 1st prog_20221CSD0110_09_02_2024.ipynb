```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
print("step 1")

```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()
x=np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x, np.sin(x), label='sin', color='blue',linestyle='dashed')
ax.plot(x, np.sin(x), label='sin x+1', color='blue',linestyle='dashed')
ax.plot(x, np.cos(x), label='cos', color='red',linestyle='dotted')

ax.set_title('multiple Lines')
ax.set_xlabel('period')
ax.set_ylabel('amplitute');
ax.legend()
plt.show()

```
